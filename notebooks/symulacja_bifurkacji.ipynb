## **Diagram bifurkacyjny - wyniki i punkty krytyczne**
Na podstawie przeprowadzonych symulacji numerycznych otrzymano diagram bifurkacyjny przedstawiający zależność $v^*$ od $a$.

```python
from bifurcation import run_bifurcation, plot_bifurcation

# Należy podać: 
# parametry modelu: m, d1, d2
# parametry przestrzeni [0, Lx] x [0, Ly] z podziałem na Nx i Ny znajdujących się w nich wartości
# krok czasowy ht
result = run_bifurcation(m = 0.45, d1 = 1.5, d2 = 0.02,
    Lx = 10, Ly = 10, Nx = 30, Ny = 30, ht = 0.025)
plot_bifurcation(result)
```

Możemy zaobserwować ciekawe zachowanie funkcji w okolicy *tipping point*.

Widzimy, że gdy startujemy z dużego $a$ (np. $a=4m$) tworzą się falę, które w realnym odwzorowaniu mają postać pętel, natomiast nie jest to możliwe do pokazania w powyższej symulacji. Miejsca skoków funkcji odwzorowuję walkę biomasy roślin o zasoby wody poprzez zbieranie się w skupiska. Zbyt małe skupisko - biomasa nie ma szans przetrwać, zbyt duże - brakuje zasobów. Stąd na wykresie obserwujemy gwałtowne skoki i spadki, ponieważ choć przez chwilę skupiska umożliwiają przedłużenie przetrwania. Gdy przypływ wody jest coraz mniejszy, skupiska wymierają. Po przekroczeniu pewnego progu dla $a<2m$ widzimy ostatni skok, po czym roślinność na danym obszarze wymiera.

Trochę inna sytuacja występujemy, gdy startujemy symulacje z $a \in (0, 2m),$ wtedy choć biomasa roślin zaczyna z niezerowej ilości, nie ma szans przetrwać i widzimy szybki spadek do 0, na którym stan biomasy już pozostaje do końca symulacji.
